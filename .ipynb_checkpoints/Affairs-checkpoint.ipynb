```python
# plot number of submitted and accepted motions per year (motion)

# select and group by year and affair type
df_year_affair_accepted = df[ (df["affair_type"] == "MOTIO") & (df["accepted"] == True) ]
df_year_affair_accepted = df_year_affair_accepted.groupby([df_year_affair_accepted.date_received.dt.year, df_year_affair_accepted.affair_type]).size().unstack()

# submitted affairs per year
df_year_affair_sub = df[ (df["affair_type"] == "MOTIO") ]
df_year_affair_sub = df_year_affair_sub.groupby([df_year_affair_sub.date_received.dt.year, df_year_affair_sub.affair_type]).size().unstack()

#rename columns
df_year_affair_sub.columns = ["MOTIO_sub"]
df_year_affair_accepted.columns = ["MOTIO_acc"]

# merge dfs
df_year_affair_def = pd.concat([df_year_affair_sub, df_year_affair_accepted,], axis=1)

print(df_year_affair_def)

# plot
df_year_affair_def.plot(kind="bar", rot=0, stacked=False, color=["C2", "grey",])

# custom legend
L=plt.legend(loc="best", prop={"size":15})
L.get_texts()[0].set_text("Eingereichte Motionen")
L.get_texts()[1].set_text("Angenommene Motionen")

# misc
plt.title("Eingereichte und angenommene Motionen \n im Obwaldner Kantonsrat pro Jahr\n(2010-2020)", fontsize="20")
plt.xlabel("Jahr", size="15")
plt.ylabel("Anzahl", size="15")

plt.savefig('Fig_Motionen.png', dpi=300, bbox_inches='tight')
```

```python
# compare relative share of submitted affairs per fraction with mean of fraction strength

# sort df by fraction (get absolute)
df_fraction_affair_999 = df[ (df["signatory_fraction_id"] != 999) ]
df_fraction_relative_affair = df_fraction_affair_999.groupby([df_fraction_affair_999.signatory_fraction]).size()

# sum of total affairs
affair_sum = df_fraction_relative_affair.sum()
# relative affairs
relative = df_fraction_relative_affair.apply(lambda x: 100*x/affair_sum)

relative_df = relative.reset_index()
# rename 0 column
relative_df_new = relative_df.rename(columns={0: "share_affairs"})
# hardcoded mean of fraction strength (2010-2020)
# order: csp, cvp, fdp, sp, svp, unknown
#strength = ["13.9393939393939", "33.3333333333333", "16.969696969697", "12.1212121212121", "23.6363636363636", "0"]
strength = ["13.9393939393939", "33.3333333333333", "16.969696969697", "12.1212121212121", "23.6363636363636",]
relative_df_new['strength'] = strength
# convert to nummeric
relative_df_new.strength = pd.to_numeric(relative_df_new.strength)

print(relative_df_new)

relative_df_new.head()

# plot
relative_df_new.plot(kind="bar", rot=0, x="signatory_fraction")
# custom legend
L=plt.legend(loc="best", prop={"size":15})
L.get_texts()[0].set_text("Anteil Vorstösse")
L.get_texts()[1].set_text("Anteil Mandate (Durchschnitt)")

# misc
plt.title("Anteil eingereichter Vorstösse und \n mittlerer Anteil Kantonsratsmandate pro Fraktion \n (2010-2020)", fontsize="20")
plt.xlabel("Fraktion", size="15")
plt.ylabel("Anteil in %", size="15")

plt.savefig('Fig_Anteil Vorstösse und Mandate.png', dpi=300, bbox_inches='tight')
```

```python
# Success AV: anon_yes

# only postulate and motion

# select from dataframe
df_success = df[ (df["affair_type"] == "MOTIO") | (df["affair_type"] == "POSTU") ]
# explore
#df_success.head()
#df_success.describe()
#df_success.info()

# ----------------------------------------------------------------------------------
# UV Anzahl Mitunterzeichnende
# ----------------------------------------------------------------------------------

# define x and y
x = pd.DataFrame(df_success["joint_signatories_count"])
y = pd.DataFrame(df_success["anon_yes"])

model = LinearRegression()
model.fit(x, y)

# plot (new with seaborn + regression line)
#df_success.plot(kind="scatter", x="joint_signatories_count", y="anon_yes")
sns.regplot(x, y)
plt.savefig('Fig_MarginalE Erfolg POSTU MOTIO joint signatories.png', dpi=300, bbox_inches='tight')


# misc

# model validation
X2 = df_success["joint_signatories_count"]
X2 = sm.add_constant(X2) # adding a constant
olsmod = sm.OLS(df_success["anon_yes"], X2).fit()
print(olsmod.summary())
```

```python
# ----------------------------------------------------------------------------------
# 2. UV Fraktion
# ----------------------------------------------------------------------------------

#new dataframe; filter out invalid
df_success_fraction = df[ ((df["affair_type"] == "MOTIO") | (df["affair_type"] == "POSTU")) ]

# drop 999 values
df_success_fraction_999 = df_success_fraction[df_success_fraction["signatory_fraction_id"] != 999]

df_success_fraction_999.head()

# define x and y
x = pd.DataFrame(df_success_fraction_999["signatory_fraction"])
y = pd.DataFrame(df_success_fraction_999["anon_yes"])

# plot
df_success_fraction_999.boxplot(by="signatory_fraction", column=["anon_yes"], grid=False)

# misc
plt.title("")
plt.suptitle("")
plt.xlabel("", size="15")
plt.ylabel("Absolute Anzahl Ja-Stimmen", size="15")

# model validation
fit = ols("anon_yes ~ C(signatory_fraction)", data=df_success_fraction_999).fit()
fit.summary()
```
